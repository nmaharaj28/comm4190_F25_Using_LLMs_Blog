## Your 3 Meal Plan for the Week

Each meal is designed to yield 2-3 servings, so you'll have leftovers for lunch or another dinner!

### Meal 1: Hearty Chicken & Spinach Pasta

*   **Ingredients Used:** Chicken, Onion, Garlic, Spinach, Canned Crushed Tomatoes, Pasta, Chicken Broth (or water), Olive Oil, Salt, Pepper, Dried Oregano.
*   **Instructions:**
    1.  Cook pasta according to package directions. Drain and set aside.
    2.  Dice chicken into bite-sized pieces. Season with salt and pepper.
    3.  Chop half the onion and mince 2-3 cloves of garlic.
    4.  In a large pot or deep skillet, heat a tablespoon of olive oil over medium-high heat. Add chicken and cook until browned and cooked through. Remove chicken and set aside.
    5.  Add a little more oil if needed, then sauté the chopped onion until softened (about 3-5 minutes). Add garlic and sauté for another minute until fragrant.
    6.  Pour in the can of crushed tomatoes and about 1/2 cup of chicken broth (or water). Add 1 tsp dried oregano, salt, and pepper to taste. Bring to a simmer.
    7.  Stir in the fresh spinach and cook until wilted.
    8.  Return the cooked chicken to the sauce. Add the cooked pasta and toss to combine. Serve warm.

### Meal 2: Quick & Easy Vegetarian Chili with Rice

*   **Ingredients Used:** Onion, Garlic, Green Bell Pepper, Canned Diced Tomatoes, Kidney Beans, Black Beans, Rice, Chili Powder, Cumin, Paprika, Salt, Pepper, Shredded Cheddar (optional).
*   **Instructions:**
    1.  Cook rice according to package directions.
    2.  Chop the remaining half onion, the green bell pepper, and mince 2-3 cloves of garlic.
    3.  In a large pot or Dutch oven, heat a tablespoon of olive oil over medium heat. Add chopped onion and bell pepper and sauté until softened (about 5-7 minutes). Add minced garlic and cook for another minute.
    4.  Drain and rinse the kidney beans and black beans.
    5.  Stir in the two cans of diced tomatoes (undrained), the drained beans, 1 tbsp chili powder, 1 tsp cumin, 1/2 tsp paprika, salt, and pepper to taste. Add about 1/2 cup of water or chicken broth if you like a thinner chili.
    6.  Bring to a simmer, then reduce heat to low, cover, and let it cook for at least 15-20 minutes for flavors to meld, stirring occasionally.
    7.  Serve hot over cooked rice, topped with shredded cheddar cheese if desired.

### Meal 3: Lemon Herb Baked Chicken with Roasted Potatoes & Broccoli

*   **Ingredients Used:** Chicken, Potatoes, Broccoli, Lemon, Garlic, Olive Oil, Salt, Pepper, Dried Oregano (or Dried Parsley/Basil if you have it).
*   **Instructions:**
    1.  Preheat oven to 400°F (200°C).
    2.  Wash and chop potatoes into 1-inch cubes. Chop broccoli into florets. Mince 2-3 cloves of garlic.
    3.  In a large bowl, toss the potatoes and broccoli with 2 tablespoons of olive oil, half the minced garlic, salt, and pepper. Spread them in a single layer on a baking sheet.
    4.  Cut the remaining chicken breasts/thighs into similar-sized pieces (or leave whole if you prefer). In the same bowl, toss the chicken with 1 tablespoon of olive oil, the remaining minced garlic, 1 tsp dried oregano, the juice of half the lemon, salt, and pepper.
    5.  Add the seasoned chicken to the baking sheet with the vegetables, ensuring everything is in a single layer.
    6.  Bake for 25-35 minutes, flipping the chicken and vegetables halfway through, until the chicken is cooked through (internal temperature 165°F/74°C) and the potatoes are tender and lightly browned.
    7.  Squeeze the remaining lemon half over the chicken and vegetables before serving.

Enjoy your delicious and budget-friendly meals from Acme!

**Tips for Single Cooking & Minimizing Waste:**

*   **Freeze Smart:** For items like chicken broth or crushed tomatoes, if you don't use the whole container/can, you can freeze the rest in smaller portions (ice cube trays work great for broth!).
*   **Get Creative with Leftovers:** If you get bored of eating the same thing, try repurposing. Leftover chicken from Meal 3 could go into a quick salad or sandwich. Leftover spinach can be added to scrambled eggs.
*   **"Use It Up" Meal:** Towards the end of the week, if you have small bits of veggies or protein left, make a "fridge clean-out" stir-fry, omelet, or soup.
*   **Meal Prep Day:** Dedicate an hour or two on one day to cook these meals. This way, you have your dinners and lunches ready to go for several days!
