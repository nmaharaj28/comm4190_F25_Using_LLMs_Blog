***Reflection***

This post was fun to create! As someone from Trinidad, I think the art based on Trinidad was accurate! It incorporated the steel pan, which is a major part of Trinidadian culture, as well as a Carnival mask, the flag, and palm trees. The art based on India was also interesting. It highlighted key parts of Indian culture, like traditional fashion, the lotus flower, and the Taj Mahal. Both pieces captured the vibrancy and spirit of their respective cultures through color and composition. I especially liked how both artworks celebrated music, tradition, and community as central themes that connect people across different cultures.
